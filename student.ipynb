```python
def raw_vals_to_categorical(df, col_name):
    
    def convert_to_int(n): #To get data into format range function will recognize
        return int(n * 10)
    
    values = df[col_name].values
    values_sorted = sorted(df[col_name].values, reverse=True)
    
    values_as_int = list(map(convert_to_int, values))
    values_sorted_as_int = list(map(convert_to_int, values_sorted))
    
    top5_cutoff = values_sorted_as_int[4]
    top10_cutoff = values_sorted_as_int[9]
    top20_cutoff = values_sorted_as_int[19]
    top30_cutoff = values_sorted_as_int[29]
    top40_cutoff = values_sorted_as_int[39]
    top50_cutoff = values_sorted_as_int[49]
    top100_cutoff = values_sorted_as_int[99]
    
    
    print(top5_cutoff)
    print(top10_cutoff)
    print(top20_cutoff)
    print(top30_cutoff)
    print(top40_cutoff)
    print(top50_cutoff)
    print(top100_cutoff)
    
    values_categorical = []
    
    for v in values_as_int:
        if v >= top5_cutoff:
            values_categorical.append('Top 5')
        elif v in range(top10_cutoff, top5_cutoff):
            values_categorical.append('Top 10')
        elif v in range(top20_cutoff, top10_cutoff):
            values_categorical.append('Top 20')
        elif v in range(top30_cutoff, top20_cutoff):
            values_categorical.append('Top 30')
        elif v in range(top40_cutoff, top30_cutoff):
            values_categorical.append('Top 40')
        elif v in range(top50_cutoff, top40_cutoff):
            values_categorical.append('Top 50')
        elif v in range(top100_cutoff, top50_cutoff):
            values_categorical.append('Top 100')
        elif v < top100_cutoff:
            values_categorical.append('Outside Top 100')
            
    return values_categorical
```

```python
df = raw_df.copy(deep=True)


columns = list(df.columns)
columns = columns[1:-1]
columns

for c in columns:
    df[c] = raw_vals_to_categorical(df,c)
    
df
```

```python
#un-nest

def get_probability_distribution(df, col_name):
    master_list = []

    def get_vals(df, col_name):
        return df[col_name].values

    def get_unique_vals(df, col_name):
        values = df[col_name].values
        return np.unique(values)
    
    vals = get_vals(df, col_name)
    unique_vals = get_unique_vals(df, col_name)
    
    def get_class_distribution(df, col_name, i):
        yes_outcomes = 0
        no_outcomes = 0
    
        for v in vals:
            if v == unique_vals[i]:
                yes_outcomes +=1
            else:
                no_outcomes +=1
        
        class_distribution = [yes_outcomes,no_outcomes]
        return class_distribution
    
    j=0
    while j < len(unique_vals):
        distribution = get_class_distribution(df, col_name, j)
        #print(unique_vals[j])
        master_list.append(distribution)
        j+=1
        
    return master_list

def entropy(pi):
    '''
    return the Entropy of a probability distribution:
    entropy(p) = - SUM (Pi * log(Pi) )
    '''

    total = 0
    for p in pi:
        p = p / sum(pi)
        if p != 0:
            total +=  p * log(p, 2)
        else:
            total += 0
    total *= -1
    return total

def information_gain(D, a, a_as_str):
    '''
    return the information gain:
    gain(D, A) = entropy(D)− SUM( |Di| / |D| * entropy(Di) )
    '''

    total = 0
    for Di in a:
        total += abs(sum(Di) / sum(D)) * entropy(Di)

    gain = abs(entropy(D) - total)
    print('information gain of ' + a_as_str + ': ' + str(total))
```
